```python
df_sample = df.sample(10000, random_state = 42)
isNumeric = df.select_dtypes(include=[np.number])
isNumeric[['member_gender', 'user_type']] = df[['member_gender', 'user_type']]
isNumeric_sample = isNumeric.sample(100000, random_state = 42)
g = sns.pairplot(isNumeric_sample[['duration_sec',  'member_birth_year','member_gender', 'user_type']], hue="member_gender", palette="Dark2",  aspect = (14.70/2)/(8.27/2))
plt.title('People born in different years vs time duration of bike usage')
```

The above graph represents the data where men ,women , and people of other categories went for bike rides. The number of women who went for bike rides seems to be more than the number of men.
> For women in general, majority went for bike rides for almost 20000 seconds.

> For men who were born in between 1920-1940, went for bike rides for almost 20000 seconds.
> For men who were born in the mid 80's and 90's , they went for bike rides for almost 40000 to 60000 seconds.
> There were women who did go for bike rides for 40000 to 60000 secs. However the numbers are less in comparison to   men. 

```python
plt.figure(figsize=[14.70,8.27])
sns.violinplot(data=df_sample[df_sample.duration_sec < df_sample.duration_sec.quantile(0.99)],y="duration_sec",hue="user_type",x="member_gender",dodge=True,pallete="RdYlBy_r")
plt.title('Customers vs Suscribers-time')
```

In  the above graph, using a violin plot we have tried to show for how long Customers and Subscribers went for bike rides. 
>A lot of Male Customers went for bike rides for about 800 seconds.
>A lot of Female Customers went for bike rides for about 900 seconds.

>In the case of subscribers, if you see Male, Female or other gender Subscribers, the violin plot is a little more skewed as compared to Customers, which suggest that there were small number of long bike trips taken by the subscribers.

```python
df['Age'] = 2019 - df['member_birth_year'] # creates age and age group column
df["Age Groups"] = pd.cut(df.Age, [0, 24, 55, 65, 100],labels = ["Youth", "Adult", "Older Adult", "Elderly"], include_lowest=True)
```

```python
sns.countplot(x = "Age Groups", data = df, color = "grey")
plt.title('Number of people in different age groups')
```

The above graph shows us how many people do we have of the various age groups.
> There are around 25000 youths, 5000 older adults, 1000 elderly people and having the maximum number of adults with 13000 number of people.

```python
sns.violinplot(data=df, y="duration_sec",x="member_gender", hue="Age Groups", dodge=True,palette="RdYlBu_r")
plt.ylim(-60, df.duration_sec.quantile(0.95)) # this limits the y values to exclude the top 5 percentile
plt.title('People of different age groups using bike rides')
```

>With the above graph we can see that young Males took bike rides for 270 secs while young females took bike rides for 280 secs, however males took more trips as compared to females. On the other hand people of other categories of the young generation took rides for 500 secs.

> This graph in particular also shows that the elderly males and males of the other categories took very less number of bike rides.

```python
df.start_time.value_counts()
```

```python
df['start_time'] = pd.to_datetime(df.start_time)
df['day']=df.start_time.dt.strftime('%A')
df['day']
```

```python
df['month'] = df.start_time.dt.strftime('%B')
df['month']
```

```python
df['hour'] = df.start_time.dt.strftime('%H')
df['hour']
```

```python
plt.subplot(1, 1, 1)
#df_customer = df.query('user_type == "Customer"').groupby([df['hour'], df['day']])['bike_id'].size().reset_index();
df_customer = df.query('user_type == "Customer"')
df_heatmap1 = df_customer.pivot_table(index='day',columns='hour', values='bike_id', aggfunc='count')
# pivot table helps in telling pandas to have categorical data that have an order like fri, sun, sat to be plotted accordingly
sns.heatmap(df_heatmap1)
plt.title('Customers hourly and weekly usage')
```

```python
plt.subplot(1, 1, 1)
df_subscriber=df.query('user_type == "Subscriber"')
df_heatmap2= df_subscriber.pivot_table(index='day',columns='hour', values='bike_id', aggfunc='count')
sns.heatmap(df_heatmap2)
plt.title('Subscribers hourly and weekly usage')
```

From the above two heatmaps  it is clear that for how many hours on what day Customers and Subscribers took bike rides.
>In the case of Subscribers, they took bike rides only on weekdays. Maximum people in them used to travel at around 6, 7 ,8,9 am or at 5, 6, 7 pm.

>In the case of Customers, they took bike rides both on weekdays and weekends.
> On saturdays and Sundays mostly their rides used to range from around 10 am to 4 pm.
